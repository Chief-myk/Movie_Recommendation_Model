```python
import ast

def extract_companies(x):
    try:
        if pd.isna(x):
            return ""

        if isinstance(x, bool):
            return ""

        if isinstance(x, str):
            x = ast.literal_eval(x)

        return " ".join(
            item['name'].replace(" ", "_")
            for item in x
            if isinstance(item, dict) and 'name' in item
        )

    except:
        return ""

df['production_companies'] = df['production_companies'].apply(extract_companies)
```

```python
def extract_countries(x):
    try:
        if pd.isna(x):
            return ""

        if isinstance(x, bool):
            return ""

        if isinstance(x, str):
            x = ast.literal_eval(x)

        return " ".join(
            item['name'].replace(" ", "_")
            for item in x
            if isinstance(item, dict) and 'name' in item
        )

    except:
        return ""

df['production_countries'] = df['production_countries'].apply(extract_countries)
```

```python
# nltk (Natural Language Toolkit)
# A popular Python library for NLP (Natural Language Processing).
# Used for text preprocessing, tokenization, stemming, lemmatization, stopword removal, etc.

import nltk

# stopwords
# Provides a list of common words like "the", "is", "and", "a", etc.
# These words usually don't add much meaning and are often removed before training NLP models.

from nltk.corpus import stopwords

# WordNetLemmatizer
# Converts words to their base/root form while preserving meaning.
# Example:
# "running" -> "run"
# "better" -> "good"
# "cars" -> "car"

from nltk.stem import WordNetLemmatizer

# re (Regular Expressions)
# Built-in Python library for text cleaning and pattern matching.
# Commonly used to:
# - Remove special characters
# - Remove numbers
# - Extract patterns (emails, URLs, etc.)
# Example:
# re.sub(r'[^a-zA-Z]', ' ', text)

import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split() # tokenization
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['original_title']).drop_duplicates()
indices
```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = [
    'popularity',
    'runtime',
    'vote_average',
    'vote_count',
    'release_year'
]

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=50000 , stop_words = "english",ngram_range=(1,2))
vectors = tfidf.fit_transform(df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(vectors[idx], vectors).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['original_title'].iloc[similar_idx]
```

```python
recommend('Apollo 13')
```
